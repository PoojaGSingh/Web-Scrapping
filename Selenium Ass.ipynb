```python
quotes=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
quotes.click()
```

```python
Quote = []
Author = []
ToQ = []

start=0
end=10

for page in range(start,end):
    #Quote
    quote=driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote:
        Quote.append(i.text)
    
    #Author
    author=driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in author:
        Author.append(i.text)
        
    #Type of Quote
    type=driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type:
        ToQ.append(i.text)
        
    
```
