```python
H_Name =[]
Dist = []
Ratings = []
T_Reviews = []
O_Reviews= []
p_price = []
d_price = []
Facilities = []
P_discrp = []

start=0
end=3

for page in range(start,end):
        
    h_name=driver.find_elements(By.XPATH, '//h2[@class="title title-6"]')
    for x in h_name:
        h = (x.text)
        if h == (""):
            H_Name.append('-')
        else:
            H_Name.append(h)
            
    dist=driver.find_elements(By.XPATH, '//span[@class="description"]')
    for x in dist:
        d = (x.text.split(" "))
        if d == (""):
            Dist.append('-')
        else:
            Dist.append(d[2])
                
    rate=driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]')
    for x in rate:
        r = (x.text.split("\n"))
        if r == (""):
            Ratings.append('-')
        else:
            Ratings.append(r[0])
        
    trw=driver.find_elements(By.XPATH ,'//div[@class="reviews"]')
    for x in trw:
        tr = (x.text.split(" "))
        if tr == (""):
            T_Reviews.append("-")
        else:
            T_Reviews.append(tr[0])
   
    
    orw=driver.find_elements(By.XPATH,'//div[@class="keyword"]//span')
    for x in orw:
        ow = (x.text)
        O_Reviews.append(ow)
                
                
    ppr=driver.find_elements(By.XPATH, '//div[@class="price-col"]')
    for x in ppr:
        pp = (x.text)
        if pp == (""):
            p_price.append("-")
        else:
            p_price.append(pp)
            
            
    dpr=driver.find_elements(By.XPATH, '//div[@class="price-col"]')
    for x in dpr:
        dp = (x.text)
        if dp == (""):
            d_price.append("-")
        else:
            d_price.append(dp)
                
    fac=driver.find_elements(By.XPATH, '//div[@class="facilities-label facilities"]')
    for x in fac:
        f = (x.text.split("\n"))
        if f == (""):
            Facilities.append("-")
        else:
            Facilities.append(f)
                
    des=driver.find_elements(By.XPATH, '//div[@class="rating-factors prop-card-tablet rating-factors small"]')
    for x in des:
        d = (x.text.split("\n"))
        if d == (""):
            P_discrp.append("-")
        else:
            P_discrp.append(d)
                
        
  
```
